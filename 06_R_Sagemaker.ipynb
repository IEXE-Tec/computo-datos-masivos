## For & Print
A diferencia de python, R no utiliza identacion para definir loops o condicionales. Estos bloques de codigo estan definicos por los parentesis `(){}`. 
+ `()` Define la parde condicional o el iterador.
+ `{}` Define el bloque de codigo que se ejecutara.

````
a = 5:10
for (i in a){
print(i)
}
````

`paste` es una funcion que concatena/pega strings. Tambien podriamos hacer uso de la funcion `sprintf` que tiene una sintaxis parecida a la forma de imprimir de python.

````
for (i in 1:length(a)){
  print(paste("El indice ", i, " del array a contiene el valor", a[i]))
}
````

Al igual que python podemos iterar sobre las listas.

````
for (i in c(1,3 , "ax")){
  print(i)
}
````

## Funciones

La sintaxis para definir una funcion es la siguiente:
``` nombre_funcion <- function(arg1, arg2,...,argn){  ... return(valor_de_retorno) }```

Definamos la funcion suma

````
suma <- function(a,b){
  s <- a + b
  return (s)
}
````

````
suma(3,4)
````

Tambien podemos omitir el keyword **return**, y por default regresara el ultimo valor/operacion calculada.

````
resta <- function(a, b){
  s <- a-b
}
````

````
z <- resta(4,3)
z
````

**Valores Default**  
Tambien podemos agregar valores por default a la funcion como lo hacemos en python.

````
resta <- function(a, b=0){
  s <- a-b
}
z <- resta(c(1,2))
z
````

## Funciones vectoriales

A diferencia de python, las funciones de R extrapolan los inputs vectoriales a funciones de manera nativa.

````
z <- suma(c(4,5), c(1,2))
print(z)
print(z[1])
print(z[2])
````

````
class(z)
````

**Que pasa si los inputs no tienen las dimensiones correctas?**

````
z <- suma(c(4,5,-1), c(1,2))
print(z)
````

Nos manda un `warning` y extrapola la dimension faltante repitiendo los valores del vector con dimension menor, respetando el orden. En nuestro caso dado que `c(1,2)` tiene una dimension menos que `c(4,5-1)` R realiza la operacion con `c(1,2,1)`. Veamos otro ejemplo.

````
z <- suma(c(4,5,-1,0), c(1,2))
print(z)
````

Lo mismo pasa cuando aplicamos operaciones a las listas, pero **no se emite el warning**.

````
z <- c(4,5,-1,0) + c(1,2)
print(z)
````

## Magrittr %>%

La libreria de magritrr (parte de tidyverse) que importamos anteriormente tiene un operador muy particular conocido como pipe `%>%`. Este fuciona muy parecido al pipe `|` que utilizamos en la terminal. Nos permite pasar una variable a la siguiente funcion como primer argumento.

````
# Recordemos como definimos la funcion de resta
resta <- function(a, b=0){
  s <- a-b
}
````

````
z <- c(1,2) %>% resta()
z
````

Si la funcion tiene mas argumentos, se seguira respetando el orden de los argumentos, y el argumento pasado por `%>%` suplira el primero.

````
z <- c(0,0) %>% resta(c(2,3))
z
````

````
# Recordemos como definimos la funcion de resta
fun_2 <- function(a, b=0, c){
  s <- c*(a-b)
}
````

Si usamos `%>%` en conjunto con los argumentos explicitos de la funcion, el argumento que pasemos por *pipe* suplira al primero vacio. Por ejemmplo, pasemos el argumento por *pipe* en nuestra funcion definida anteriormente.

````
z <- c(0) %>% fun_2(a=1,c=2)
z
````

La sintaxis, y estructura de **Tidyverse** esta pensada para permitir al usuario expresar el flujo de codigo con base en el perador `%>%`. Casi cualquier codigo de ciencia de datos, lo utilizara.
